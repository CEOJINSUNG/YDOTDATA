```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

### CSS Selector
- name : section > div.ranking-wrap > div:nth-child(2) > div > div > div.ranking_cont > ul.list_ranking > li:nth-child(1) > div.ranking_info > div.ranking_txt > span.title
- sub : section > div.ranking-wrap > div:nth-child(2) > div > div > div.ranking_cont > ul.list_ranking > li:nth-child(1) > div.ranking_info_right > div.ranking_data > ul > li:nth-child(1) > span
- genre : section > div.ranking-wrap > div:nth-child(2) > div > div > div.ranking_cont > ul.list_ranking > li:nth-child(1) > div.ranking_info > div.ranking_txt > span.category

```python
page = range(1, 24)
url = "https://socialerus.com/Ranking/Index?PageNo={0}&ChCategory=&ChTheme=&ChSubscMin=0&ChSubscMax=1000&ChViewMin=0&ChViewMax=18404994235&ChVideoMin=0,95664&Sort=subsc"
url_list = []
for r in page:
    url_list.append(url.format(r))
print(url_list)
```

```python
tenyoutuber = []
people = range(1, 21)
for a in url_list:
    req = requests.get(a)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    
    for person in people:
        name = soup.select("section > div.ranking-wrap > div:nth-child(2) > div > div > div.ranking_cont > ul.list_ranking > li:nth-child({0}) > div.ranking_info > div.ranking_txt > span.title".format(person))
        sub = soup.select("section > div.ranking-wrap > div:nth-child(2) > div > div > div.ranking_cont > ul.list_ranking > li:nth-child({0}) > div.ranking_info_right > div.ranking_data > ul > li:nth-child(1) > span".format(person))
        genre = soup.select("section > div.ranking-wrap > div:nth-child(2) > div > div > div.ranking_cont > ul.list_ranking > li:nth-child({0}) > div.ranking_info > div.ranking_txt > span.category".format(person))
        for item in zip(name, sub, genre):
            tenyoutuber.append(
                {
                    "name" : item[0].text,
                    "sub" : item[1].text,
                    "genre" : item[2].text
                }
            )
        
data = pd.DataFrame(tenyoutuber)
data.to_csv("TenSocial.csv")
```
