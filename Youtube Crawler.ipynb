```python
def getData(number):
    init = pd.DataFrame([])
    for i in number:
        url = "https://www.musicow.com/song/{0}?tab=price".format(i)
        req = requests.get(url)
        # req.text하면 그 사이트의 html 코드를 가져올 수 있다.
        html = req.text
        # 가져온 html 코드를 BeautifulSoup(html, 'html.parser')를 활용해 깔끔하게 정리
        soup = BeautifulSoup(html, 'html.parser')
        listdata = []
        for j in range(1, 16):
            date = str(soup.select("div.market_table > table > tbody > tr:nth-child({0}) > td:nth-child(1)".format(j))[0]).replace("<td>", "").replace("</td>", "")[0: 8]
            price = str(soup.select("div.market_table > table > tbody > tr:nth-child({0}) > td:nth-child(2)".format(j))[0]).replace("<td>", "").replace("</td>", "")
            amount = str(soup.select("div.market_table > table > tbody > tr:nth-child({0}) > td:nth-child(3)".format(j))[0]).replace("<td>", "").replace("</td>", "")
            listdata.append([date, price, amount])
        ex = pd.DataFrame(listdata, columns=["date", "price", "amount"])
        init = pd.concat([init, ex])

    return init
getData(number)
```

### 여기서 가져올 정보
1. 한국 유튜버(name) - list-skin > form:nth-child(4) > table > tbody > tr:nth-child(1) > td.subject > h1 > a
2. 한국 유튜버 구독자 수(sub) - list-skin > form:nth-child(4) > table > tbody > tr:nth-child(1) > td.subscriber_cnt
3. 한국 유튜버 장르(genre) - list-skin > form:nth-child(4) > table > tbody > tr:nth-child(1) > td.subject > h1 > p

### 참고
- soup.select는 선택된 태그를 보여준다
- [0]은 선택된 태그의 개수가 여러개일 경우 첫번째 요소를 가리킴
- text는 선택된 태그의 요소를 가져옴
- replace('\n', '').replace('\t', '').replace('  ', '')는 앞뒤 공백 띄어쓰기 제거

```python
name = soup.select("table > tbody > tr:nth-child(1) > td.subject > h1 > a")
print(name)
```

```python
sub = soup.select("table > tbody > tr:nth-child(1) > td.subscriber_cnt")
print(sub)
```

```python
genre = soup.select("table > tbody > tr:nth-child(1) > td.subject > h1 > p")
print(genre)
```

```python
# youtube 배열
youtuber = []
for item in zip(name, sub, genre):
    youtuber.append(
        {
            "name" : item[0].text.replace('\n', '').replace('\t', '').replace('  ', '').replace('\r', ''),
            "sub" : item[1].text.replace('\n', '').replace('\t', '').replace('  ', ''),
            "genre" : item[2].text.replace('\n', '').replace('\t', '').replace('  ', ''),
        }
    )
print(youtuber)
```

### 페이지 구조
- 한 페이지에 100명의 유튜버가 존재
- #list-skin > form:nth-child(4) > table > tbody > tr:nth-child(2) > td.subject > h1 > a
- tr:nth-child(2)이 1에서 100까지 존재
- people = range(1, 101)
- 구독자 '비공개'를 제외하고 53페이지까지 존재
- page = range(1, 54)

### 표준 변수 선언
- name : "유튜버 이름"
- sub : "구독자 수"
- genre : "장르"

```python
page = range(1, 54)
url = "https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page="
url_list = []
for r in page:
    url_list.append(url+str(r))
print(url_list)
```

```python
youtuber = []
people = range(1, 101)
for url in url_list:
    req = requests.get(url)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    
    for person in people:
        name = soup.select("table > tbody > tr:nth-child({0}) > td.subject > h1 > a".format(person))
        sub = soup.select("table > tbody > tr:nth-child({0}) > td.subscriber_cnt".format(person))
        genre = soup.select("table > tbody > tr:nth-child(1) > td.subject > h1 > p".format(person))
        for item in zip(name, sub, genre):
            youtuber.append(
                {
                    "name" : item[0].text.replace('\n', '').replace('\t', '').replace('  ', '').replace('\r', ''),
                    "sub" : item[1].text.replace('\n', '').replace('\t', '').replace('  ', ''),
                    "genre" : item[2].text.replace('\n', '').replace('\t', '').replace('  ', ''),
                }
            )
            
data = pd.DataFrame(youtuber)
data.to_csv("KoreanYoutuber.csv")
```
